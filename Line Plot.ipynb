```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set()
```

# Creating a Graphic Object

```python
fig = plt.figure()
ax = plt.axes()
plt.show()
```

```python
x = np.linspace(0,10)
```

```python
y = np.sin(x)
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y)
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.plot(y, np.cos(y))
plt.show()
```

# Specifying Line Colors

```python
plt.plot(x, np.sin(x - 0), color = 'blue')
plt.show()
```
