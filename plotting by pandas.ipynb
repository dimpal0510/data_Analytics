```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
plt.style.use("fivethirtyeight")
```

# PLotting grapg with plot() method

```python
data =  pd.Series(np.random.randn(1000).cumsum())
data.plot()
plt.show()
```

```python
df1 = pd.DataFrame(np.random.randn(100,4),columns = list('ABCD'))
```

```python
df1 = df1.cumsum() # use to sum of random numbers
```

```python
df1.plot()
plt.show()
```

# Bar Charts

```python
df1.iloc[10].plot(kind = 'bar')
plt.show()
```

```python
df1.iloc[10].plot.bar()
plt.show()
```

```python
df2 = pd.DataFrame(np.random.rand(7,3), columns = list('ABC'))
df2.plot.bar()
plt.show()
```

```python
df2.plot.bar(stacked = True)
plt.show()
```

# Histograms

```python
iris = pd.read_csv("iris.data",header = None)
iris.columns = ["sepal_length","sepal_width","petal_length",
                "petal_width","species"]
iris.head()
```

```python
iris.plot.hist(alpha = 0.7)
```

```python
plt.show()
```

```python
iris.plot.hist(alpha = 1, stacked = True)
```

```python
plt.show()
```

```python
iris.hist(color = "blue", alpha = 1, bins = 20)
```

```python
plt.show()
```
