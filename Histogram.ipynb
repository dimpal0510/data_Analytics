# How to plot histograms?

```python
data = np.random.randn(1000)
plt.hist(data)
plt.show()
```

```python
plt.hist(data)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.show()
```

```python
plt.hist(data, edgecolor = "black")
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.show()
```

```python
bins = [-3,-2,-1,0,1,2,3]
```

```python
plt.hist(data, edgecolor = "black" , bins = bins)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.show()
```

```python
median = np.median(data)
```

```python
plt.hist(data, edgecolor = "black", bins = bins)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.axvline(median, color = "red")
plt.show()
```

```python
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-3,2,1000)
x3 = np.random.normal(2,1,1000)
```

```python
plt.hist(x1, histtype = "stepfilled",
         alpha = 0.5, bins = 40)
plt.hist(x2, histtype = "stepfilled",
         alpha = 0.5, bins = 40)
plt.hist(x3, histtype = "stepfilled",
         alpha = 0.5, bins = 40)
```

```python
plt.show()
```

# Two diemensional Histogram

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean, cov, 1000).T
```

```python
plt.hist2d(x,y,bins = 30, cmap = "Reds")
```

```python
plt.show()
```

```python
plt.hist2d(x,y, bins = 30, cmap = "Reds")
cb = plt.colorbar()
cb.set_label("Number of rectangles")
plt.show()
```
