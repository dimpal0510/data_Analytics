```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
fig = plt.figure()
ax = plt.axes()
plt.show()
```

# Scatter Plot with Plot Method

```python
plt.plot(iris.sepal_length, iris.sepal_width,'o')
plt.show()
```

```python
plt.plot(iris.sepal_length, iris.sepal_width,'+')
plt.show()
```

```python
plt.plot(iris.sepal_length, iris.sepal_width,'v')
plt.show()
```

```python
plt.plot(iris.sepal_length, iris.sepal_width,'rv')
plt.show()
```

```python
plt.plot(iris.sepal_length, iris.sepal_width,'rv', markersize = 10)
plt.show()
```
