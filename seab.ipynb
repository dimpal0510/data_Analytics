```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('classic')
```

```python
%matplotlib inline
```

```python
rng = np.random.RandomState(0)
```

```python
x = np.linspace(0,10,250)
```

```python
y = np.cumsum(rng.randn(250,6),0)
```

```python
plt.plot(x,y)
plt.legend('ABCDEF', ncol=2, loc='best')
plt.show()
```

```python
import seaborn as sns
```
