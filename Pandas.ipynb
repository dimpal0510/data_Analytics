```python
def extractLastName(name):
    token = name.split(',')
    return token[0]

# Map the function to the none column and assign a new column in our dataframe called Last Name

data['Last Name'] = data['Name'].apply(extractLastName)

#Let's have a look at the first 5 rows

data.loc[:4, ['Name','Last Name']]
```

```python
# Encode male as 1 and Female as 0

data['Encoded sex'] = data['Sex'].map({'male':1, 'female':0})

# Show the first 5 rows of sex and Encode sex

data.loc[:4,['Sex','Encoded sex']]
```

```python
import pandas as pd
data = pd.read_csv("gun.csv")
data
```

```python
data = data.iloc[:,:7]
data.head()
```

```python
data.drop(columns = ['incident_id','city_or_county','address'], inplace = True)
data.head()
```

```python
# Import Llibrary for datetime
import pandas as pd
from datetime import datetime

data['date'] = pd.to_datetime(data['date'], format='%d-%m-%Y') # convert date to datetime format
data['year'] = data['date'].dt.year # extract year
data['month'] = data['date'].dt.month #extract month
data['day'] = data['date'].dt.day # extract day
data['day_of_week'] = data['date'].dt.dayofweek # extract day of week
data['total_victim'] = data['n_killed'] + data['n_injured'] # sum up number of deaths and injured
data.head()
```

```python
import pandas as pd
from datetime import datetime

data['date'] = pd.to_datetime(data['date']) 
```

```python
# replace numeric day of week values with text

data['day_of_week'] = data['day_of_week'].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5 :'Saturday', 6: 'sunday'})
data.head()
```

```python
#check for null values

data.isnull().sum()
```

```python
data['day_of_week'].value_counts()
```

```python
#For count it wouldn't matter much which column we apply the function because we are only counting the number of rows
# I have used the total_victim column column here
# The order is di9fferent here because in value counts it is sorted from highest to lowest but the numbers should be indetical

data.groupby('day_of_week')['total_victim'].count()
```

```python
# What are the total number of victims from gun violence for each us state?
# we first group the number of victim by state then sum them all up

data.groupby('state')['total_victim'].sum()
```

```python
# For mean we hve to first  import Numpy

import numpy as np
import warnings
warnings.filterwarnings('ignore')
data.groupby('month')['total_victim'].agg([np.mean, min, max])
```

```python
# Drop column with misswing values

col_with_na_dropped = data.dropna(axis = 1)
col_with_na_dropped.head()
```

```python
# How much data did you loose?

print("NUmber of column in original dataset :", data.shape[1])
print("Number of column left after dropping :", col_with_na_dropped.shape[1])
difference = data.shape[1] - col_with_na_dropped.shape[1]
print("We have dropped a total of %d columns. " %difference)
```
